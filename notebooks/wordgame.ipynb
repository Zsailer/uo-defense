```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from gpmap import GenotypePhenotypeMap
from gpmap.utils import binary_mutations_map, mutations_to_genotypes
from gpgraph import GenotypePhenotypeGraph
from gpgraph.pyplot import draw_map

plt.rcParams['font.family'] = 'Arial'
```

```python
wildtype = 'GENE'
mutant = 'WORD'
mutations = binary_mutations_map(wildtype, mutant)
genotypes = mutations_to_genotypes(wildtype, mutations)
phenotypes = [
    -3.5,
    -3.75,
    -4.9,
    -6.1,
    -4.5,
    -6.3,
    -5.0,
    -4.2,
    -3.4,
    -3.0,
    -3.75,
    -5.75,
    -4.5,
    -2.9,
    -6.8,
    -2.5
]

gpm = GenotypePhenotypeMap(
    wildtype, 
    genotypes, 
    phenotypes=phenotypes,
    mutations=mutations)

G = GenotypePhenotypeGraph(gpm)
```

```python
fig, ax = plt.subplots(figsize=(5,3.75))

draw_map(
    G,
    ax=ax,
)
fig.savefig("wordgame.svg", format='svg')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from gpmap import GenotypePhenotypeMap
from gpmap.simulate import MountFujiSimulation
from gpgraph import GenotypePhenotypeGraph
from gpgraph.pyplot import draw_map
```

```python
gpm = MountFujiSimulation.from_length(6, roughness_width=0.5)
```

```python
G = GenotypePhenotypeGraph(gpm)

fig, ax = plt.subplots(figsize=(7, 3.75))
draw_map(
    G, 
    ax=ax, 
    with_labels=False, 
    cmap='viridis',
    node_size=150,
    width=1.5,
)
fig.savefig('title-network.svg', format='svg')
```
